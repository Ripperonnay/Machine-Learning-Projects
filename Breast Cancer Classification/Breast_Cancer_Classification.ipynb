**Types Of Tumor**

1. Benign (Non-Cancerous)
2. Malignant (Cancerous)

**Work Flow**
Dataset -> Data pre processing -> Train Test Split -> Logistic Regression Model

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
tumor_data = pd.read_csv('/content/data.csv')

#can also be just imported by using sklearn.datsets library [This loads data in numpy array -> Load into a pandas dataframe]
#tumor_data = sklearn.datasets.load_breast_cancer()
```

```python
#Encoding the targets to numerical value
tumor_data.replace({"diagnosis" : {"M" : 0, "B" : 1}}, inplace = True)
```

```python
tumor_data['diagnosis'].value_counts()
```

```python
tumor_data.groupby('diagnosis').mean()
```

Seperating Features and Target

```python
#Removing unwanted columns

X = tumor_data.drop(columns = ['id', 'diagnosis', 'Unnamed: 32'])
Y = tumor_data['diagnosis']
```

```python
print(X, Y)
```

###Seperating data into train and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

##Training our Model

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

```python
train_prediction = model.predict(X_train)
train_accuracy = accuracy_score(train_prediction, Y_train)
print("Accuracy on Training Data = ", train_accuracy)
```

```python
test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(test_prediction, Y_test)
print("Accuracy on Testing Data = ", test_accuracy)
```
